## 第四章：二体轨道初值问题
#### 4.1 拉格朗日系数
状态转移矩阵：
$$ \mathbf{\Phi} = \begin{bmatrix}
    F & G \\
    F_t & G_t
\end{bmatrix} $$
$$ \begin{bmatrix}
    \mathbf{r} \\
    \mathbf{v}
\end{bmatrix} = \mathbf{\Phi} \begin{bmatrix}
    \mathbf{r}_0 \\
    \mathbf{v}_0
\end{bmatrix} $$
性质：
1.状态转移矩阵的特征值为1：
$$|\mathbf{\Phi}|= {F}{G_t}-{F_t}{G} = 1$$
2.转台转移矩阵的逆：
$$ \mathbf{\Phi}^{-1} = \frac{1}{{F}{G_t}-{F_t}{G}} \begin{bmatrix}
    G_t & -G \\
    -F_t & F
\end{bmatrix} =\begin{bmatrix}
    G_t & -G \\
    -F_t & F
\end{bmatrix} $$ 
3.状态转移矩阵具有时间传递性：
$$ \mathbf{\Phi}_{0,2} = \mathbf{\Phi}_{0,1} \mathbf{\Phi}_{1,2} $$

#### 4.2 利用普适变量求解二体轨道初值问题

普适变量$\mathbf{\chi}$：
$$ \sqrt{\mu}\frac{{d{t}}}{{d{\mathbf{\chi}}}} = r $$
$\mathbf{\chi}$看做一种广义的近点角

定义几个变量：
    $$ \sigma = \frac{{\mathbf{r} \cdot \mathbf{v}}}{\sqrt{\mu}} $$
    $$ \alpha = \frac{1}{a} $$
由活力公式：
    $$ \frac{v^2}{2}  - {\frac{\mu}{r}} = {\epsilon} =  -\frac{\mu}{2a} $$
则：
$$ \alpha = {\frac{2}{r}}-\frac{v^2}{\mu} $$
将微分方程写成普适变量$\mathbf{\chi}$的方程：
$$\begin{cases} 
 \frac{d^2{\sigma}}{{d{\chi}}^2} + \alpha\sigma = 0 \\ 
 \frac{d^3{r}}{{d{\chi}}^3} + \alpha\frac{d{r}}{{d{\chi}}} = 0 \\
 \frac{d^4{t}}{d{\chi}^4} + \alpha\frac{d^2{t}}{d{\chi}^2} = 0 \\
 \end{cases} $$
依据此求微分方程的级数解。
以方程$\frac{d^2{\sigma}}{{d{\chi}}^2} + \alpha\sigma = 0$为例，
设其级数解为：
$$ \sigma = a_0(1 - \frac{\alpha\chi^2}{2!} + \frac{(\alpha\chi^2)^2}{4!} - \cdots) +a_1\chi(1 - \frac{\alpha\chi^2}{3!} + \frac{(\alpha\chi^2)^3}{5!} - \cdots) $$
$a_0,a_1$为待定系数。
令
$$ U_0(\alpha;\chi) = 1 - \frac{\alpha\chi^2}{2!} + \frac{(\alpha\chi^2)^2}{4!} - \cdots $$
$$ U_1(\alpha;\chi) = \chi(1 - \frac{\alpha\chi^2}{3!} + \frac{(\alpha\chi^2)^2}{5!} - \cdots) $$
归纳法易得：
$$ U_n(\alpha;\chi) = \chi^n(\frac{1}{n!} - \frac{(\alpha\chi^2)^{1}}{(n+2)!} + \frac{(\alpha\chi^2)^{2}}{(n+4)!} - \cdots) $$
$U_n(\alpha;\chi)$有如下递推关系：
$$
\frac{d{U_{n+1}}}{d{\chi}} = {U_{n}}
$$
$$ U_{n} + \alpha U_{n+2} = \frac{\chi^n}{n!} $$

#### 4.3圆锥曲线运动的普适函数解
##### 4.3.1 求解$\sigma$的级数解
$$ \sigma = \sigma_0{U_0} + (1-\alpha{r_0}){U_1} $$
##### 4.3.2 求解$r$的级数解
$$ r = r_0{U_0} + \sigma_0{U_1} + {U_2} $$
##### 4.3.3 求解$t$的级数解
$$ \sqrt{\mu}(t-t_0) = r_0{U_1} + \sigma_0{U_2} +{U_3} $$
##### 4.3.4 求解$\mathbf{\Phi}$的级数解
$$\begin{cases}
F = 1 - \frac{U_2(\alpha;\Chi)}{r_0} \\
G = \frac{r_0}{\sqrt{\mu}}{U_1(\alpha;\Chi)} + \frac{\sigma_0}{\sqrt{\mu}}{U_2(\alpha;\Chi)} \\ 
F_t = -\frac{\mu}{{r}{r_0}}{U_1(\alpha;\Chi)} \\
G_t = 1 - \frac{1}{r_0}{U_2(\alpha;\Chi)}
\end{cases} $$

具体求解方法：
(1)先通过广义开普勒方程
$$ \sqrt{\mu}(t-t_0) = r_0{U_1} + \sigma_0{U_2} +{U_3} $$
迭代求解$\Chi$,得到普适函数$U_n(\alpha;\chi) \space\space , n = 0,1,2,3$ 的值(只需要求出$U_0(\alpha;\chi),U_1(\alpha;\chi)$，其他可以通过递推关系得到);

(2)然后带入4.1.1~4.1.3,求解变量$\sigma$,$r$,

(3)最后带入4.3.4,求解$\mathbf{\Phi}$的级数解。

#### 4.4 求解广义开普勒方程的方法$$ \sqrt{\mu}(t-t_0) = r_0{U_1} + \sigma_0{U_2} +{U_3} $$
牛顿法的原理 ：
$$ x_{k+1} = x_{k} - \frac{f(x_k)}{f'(x_k)} $$
依照此原理可给求解广义开普勒方程的迭代公式：
$$ \chi_{k+1} = \chi_{k} - \frac{r_0{U_1(\chi_k)} + \sigma_0{U_2(\chi_k)} +{U_3(\chi_k)} -\sqrt{\mu}(t-t_0) }{r_0{U_0(\chi_k)} + \sigma_0{U_1(\chi_k)} +{U_2(\chi_k)}} $$
$\sigma$这个值有单位，很操蛋

初值设置
$$\chi_0 = \frac{\sqrt{\mu}}{r_0}{\Delta{t}} $$




```python
#导包放在一起
import numpy as np
import math
#归一化算法存在问题，以下算法均不采用任何归一化做法
```

```python
def U_0(alpha,chi):
    ans = 0.0;
    eps = 1e-10;
    for i in range(0,100):
        delta= ((alpha*(chi**2))**i)/math.factorial(2*i)
        if abs(delta) < eps:
            break
        else:
            if i%2 == 0:
                ans += delta
            else:
                ans -= delta
    return ans

def U_1(alpha,chi):
    ans = 0.0;
    eps = 1e-10;
    for i in range(0,100):
        delta= chi*((alpha*(chi**2))**i)/math.factorial(2*i+1)
        if abs(delta) < eps:
            break
        else:
            if i%2 == 0:
                ans += delta
            else:
                ans -= delta
    return ans

def U_2(alpha,chi):
    return (1 - U_0(alpha,chi))/alpha

def U_3(alpha,chi):
    return (chi - U_1(alpha,chi))/alpha


#Newton迭代法求解chi，注意此处引力参数mu归一化处理，alpha也应该归一化
def Newton(alpha,chi,r_0,sigma_0,delta_t,mu):
    for i in range(0,10000):
        chi_new = chi - (r_0*U_1(alpha,chi) + sigma_0*U_2(alpha,chi) + U_3(alpha,chi) - delta_t*np.sqrt(mu))/(r_0*U_0(alpha,chi) + sigma_0*U_1(alpha,chi) + U_2(alpha,chi))
        if abs(chi_new - chi) < 1e-10:
            break
        else:
            chi = chi_new
    return chi

#求解轨道六根数，同上一章计算结果
def root(r0_vec,v0_vec):
    mu = 398600.4418 #km^3/s^2

    h_vec = np.cross(r0_vec,v0_vec)

    print("轨道角动量为（单位：km^2/s）：",h_vec)

    h = np.linalg.norm(h_vec)
    print("轨道角动量h为：",h)

    #计算轨道倾角i
    i = np.arccos(h_vec[2]/h)
    print("轨道倾角i为（单位：°）：",i*180/np.pi)

    #计算升交点赤经Ω
    Omega = np.arctan2(h_vec[0],-h_vec[1])
    print("升交点赤经Ω为（单位：°）：",Omega*180/np.pi)

    #计算离心率e
    e_vec = np.cross(v0_vec,h_vec)/mu - r0_vec/np.linalg.norm(r0_vec)
    e = np.linalg.norm(e_vec)
    print("离心率e为：",e)

    #近心点角距
    omega = np.arctan2((e_vec[-1]),((e_vec[0]*np.cos(Omega)+e_vec[1]*np.sin(Omega))*np.sin(i)))
    print("近心点角距ω为（单位：°）：",omega*180/np.pi)

    #计算半长轴长度
    a = h**2/(mu*(1-e**2))
    print("半长轴长度a为（单位：km）：",a)

    #计算升交点角距u
    u = np.arctan2((r0_vec[-1]),((r0_vec[0]*np.cos(Omega)+r0_vec[1]*np.sin(Omega))*np.sin(i)))
    print("升交点角距u为（单位：°）：",u*180/np.pi)

    #计算近心点角距
    f = u - omega
    print("近心点角距f为（单位：°）：",f*180/np.pi)

    return h_vec,h,i,Omega,e_vec,e,omega,a,f

#求解sigma:
def get_sigma(alpha,chi,r_0,sigma_0):
    return (sigma_0*U_0(alpha,chi) + (1-alpha*r_0)*U_1(alpha,chi))

#求解r:
def get_r(alpha,chi,r_0,sigma_0):
    return (r_0*U_0(alpha,chi) + sigma_0*U_1(alpha,chi) + U_2(alpha,chi))

#求解拉格朗日状态转移矩阵：
def get_Phi(alpha,chi,r_0,sigma_0,mu,r):
    F = 1-U_2(alpha,chi)/r_0
    G = (r_0/np.sqrt(mu))*U_1(alpha,chi) + (sigma_0/np.sqrt(mu))*U_2(alpha,chi)
    F_t = -(np.sqrt(mu)/(r_0*r))*U_1(alpha,chi)
    G_t = 1 - U_2(alpha,chi)/r
    print("拉格朗日状态转移函数F为：",F)
    print("拉格朗日状态转移函数G为：",G)
    print("拉格朗日状态转移函数F_t为：",F_t)
    print("拉格朗日状态转移函数G_t为：",G_t)
    return F,G,F_t,G_t

def solve_initial_value(F,G,F_t,G_t,r0_vec,v0_vec):
    r_vec = F*r0_vec + G*v0_vec
    v_vec = F_t*r0_vec + G_t*v0_vec
    return r_vec,v_vec

def get_initial_value():
    x,y,z = map(float,input("请输入初始的位置向量(x,y,z)(单位：km)：").split())
    vx,vy,vz = map(float,input("请输入初始的速度向量(vx,vy,vz)(单位：m/s)：").split())
    delta_t = float(input("请输入时间间隔delta_t（单位：s）："))

    r0_vec = np.array([x,y,z])
    v0_vec = np.array([vx,vy,vz])/1000#单位转化为km/s
    return r0_vec,v0_vec,delta_t
```

```python
# r0_vec,v0_vec,delta_t = get_initial_value()
r0_vec = np.array([4330.155,6245.241,4387.607])
v0_vec = np.array([-5670.051,1686.977,3524.511])/1000#单位转化为km/s
delta_t = 1000#单位：s
mu = 398600.4418#km^3/s^2

h_vec,h,i,Omega,e_vec,e,omega,a,f = root(r0_vec,v0_vec)

chi_0 =(np.sqrt(mu)/np.linalg.norm(r0_vec))*delta_t

alpha = 1/a

sigma_0 = (r0_vec@v0_vec)/np.sqrt(mu)
print("sigma_0为：",sigma_0)
r0_norm = np.linalg.norm(r0_vec)

chi = Newton(alpha,chi_0,r0_norm,sigma_0,delta_t,mu)

print("牛顿法迭代结果为：",chi)
r = get_r(alpha,chi,r0_norm,sigma_0)
print("轨道半径r为：",r)
sigma = get_sigma(chi,alpha,r0_norm,sigma_0)
print("sigma为：",sigma)

F,G,F_t,G_t = get_Phi(alpha,chi,r0_norm,sigma_0,mu,r)
r_vec,v_vec = solve_initial_value(F,G,F_t,G_t,r0_vec,v0_vec)

print("初值问题的解为：")
print("位置向量r为（单位：km）：",r_vec)
print("速度向量v为（单位：km/s）：",v_vec)
```
